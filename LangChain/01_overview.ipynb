# Why LangChain?

### LangChain is an open-source framework that helps developers build applications powered by Large Language Models (LLMs).

Instead of writing everything from scratch, LangChain provides reusable components for:

- Prompt management
- LLM integration
- Memory
- RAG (Retrieval-Augmented Generation)
- Agents
- Tool calling
- Workflows
- Document processing
- Multi-agent systems

# LangChain vs SDK

An SDK is the official library provided by an LLM vendor, such as Gemini, OpenAI, Anthropic, or AWS Bedrock. It allows us to interact directly with the model for tasks like text generation, embeddings, streaming, and tool calling.

LangChain is an application framework built on top of these SDKs. It provides higher-level abstractions for building AI applications, such as prompt management, RAG pipelines, document loaders, retrievers, memory, tools, agents, and workflow orchestration.

In simple terms, an SDK helps me communicate with an LLM, while LangChain helps me build complete AI applications around that LLM.

# LangChain Architecture

LangChain follows a modular architecture where each component has a specific responsibility. User input is first processed through a prompt template, then passed through a chain that orchestrates the workflow. If external knowledge is required, a retriever fetches relevant documents from a vector database. The system can also invoke tools such as APIs or databases. Conversation context can be maintained through memory or state management. The processed prompt is sent to the LLM, and the response is structured using output parsers before being returned to the user. This modular architecture makes it easy to build RAG systems, agents, chatbots, and enterprise AI applications.

# LangChain vs LangGraph


### LangChain

Best for:

- RAG
- Simple agents
- Tool calling
- Document processing
- Chatbots

### LangGraph

LangGraph is built by the same team and is designed for:

- Stateful agents
- Multi-agent systems
- Human-in-the-loop workflows
- Complex decision trees
- Long-running AI processes

Most modern production agent systems use LangGraph underneath.


Think of it like:

- LangChain = Components
- LangGraph = Orchestration Engine
